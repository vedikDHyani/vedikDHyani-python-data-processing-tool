```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Create CSV data directly inside the notebook
csv_data = """
index,value1,value2
1,10,20
2,15,25
3,14,27
4,9,19
5,22,30
6,18,29
7,16,23
8,20,32
"""

# Load CSV from string
df = pd.read_csv(StringIO(csv_data))

# Show the DataFrame
df



# Summary statistics
print("=== Summary Statistics ===")
print(df.describe())

# Correlation
corr = df['value1'].corr(df['value2'])
print(f"\nCorrelation between value1 and value2: {corr:.3f}")



plt.figure(figsize=(8,5))
plt.plot(df["index"], df["value1"], marker="o", label="Value 1")
plt.plot(df["index"], df["value2"], marker="o", label="Value 2")
plt.grid(True)
plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Data Visualization")
plt.legend()
plt.show()


```
